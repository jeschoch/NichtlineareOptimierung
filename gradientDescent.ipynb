```python
import numpy as np
import matplotlib.pyplot as plt

x_0 = np.ones([100,2])
alpha = 0.1

def f_x(x_1,x_2): return np.power(x_1,2)+x_1-x_1*x_2+np.power(x_2,2) 

def grad_f_0(x_1, x_2): return 2*x_1-x_2+1
def grad_f_1(x_1, x_2): return -x_1+2*x_2 
```

```python
nx, ny = (100, 100)
x = np.linspace(-2, 1, nx)
y = np.linspace(-2, 1, ny)
xv, yv = np.meshgrid(x, y)
z = f_x(xv,yv)

ax = plt.figure().add_subplot(projection='3d')

ax.plot_surface(xv, yv, z, cmap='hsv', cstride=2, rstride=2)

ax.set_xlabel("x_1")
ax.set_ylabel("x_2")
ax.set_zlabel("Z label")
ax.set_zlim(0, 2)

plt.show()
```

```python
n = 100
for i in range(1,n):
    x_i = x_0[i-1,]
    x_0[i,0] = x_0[i-1,0]-alpha*grad_f_0(x_0[i-1,0], x_0[i-1,1])
    x_0[i,1] = x_0[i-1,1]-alpha*grad_f_1(x_0[i-1,0], x_0[i-1,1])
```
